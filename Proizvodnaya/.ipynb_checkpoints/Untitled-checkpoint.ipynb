### Производная

```python
from scipy.misc import derivative
```

##### scipy.misc.derivative(func, x0, dx=1.0, n=1, args=(), order=3)

```python
# первая производная (по умолчанию)
def f(x):
    return x ** 5 + 4 * np.sin(2*x) + np.cos(3*x +3)
derivative(f, 1.0, dx=1e-6)
```

```python
# вторая производная
def f(x):
    return x ** 5 + 4 * np.sin(2*x) + np.cos(3*x +3)
derivative(f, 1.0, n = 2, dx=1e-6)
```

```python
# первая производная (по умолчанию)
def f(x):
    return (np.sin(2*x + 1)) ** 5

x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, n=1, dx=1e-6) for xx in x]

plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```

```python
# вторая производная
def f(x):
    return (np.sin(2*x + 1)) ** 5

x = np.linspace(-5, 5)
fx_ = f1x
f2x = [derivative(f, xx, n=2, dx=1e-6) for xx in x]

plt.plot(x, fx_, label='function')
plt.plot(x, f2x, label='derivative')
plt.legend()
plt.show()
```
