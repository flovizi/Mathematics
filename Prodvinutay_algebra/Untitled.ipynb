### Собственные значения и собственные вектора матрицы

```python
# Найдем собственные значения и собственные вектора матрицы:

m = np.array([[2, 2],
              [1, 3]])
m
```

Функция linalg.eig() вычисляет собственные числа (значения) и собственные векторы квадратной матрицы.

Собственное число квадратной матрицы a и ее собственный вектор связаны равенством np.dot(a, v) = w*v.

```python
w, v = np.linalg.eig(m)
```

```python
# собственные значения
print(w)
```

```python
# собственные вектора
print(v)
```

```python
for i in range(len(w)):
    print("Собственное значение " + str(w[i]))
    print("Соответствующий ему собственный вектор " + str(v[:,i]))
    print()
```

### Спектральное разложение матрицы

```python
l, U = np.linalg.eig(M)
```

```python
print("Собственные значения:")
print(l)
print("\nСобственные вектора:")
print(U)
```

```python
# сформируем диагональную матрицу из собственных значений матрицы M
D = np.diag(l)
D
```

```python
# получим две ортогональные матрицы из собственных векторов матрицы M
S = U.T
print(S)
print('\n', S.T)
```

```python
# само спектральное разложение матрицы M
np.dot(np.dot(S.T, D), S)
```

#### sklearn.decomposition.TruncatedSVD 
class sklearn.decomposition.TruncatedSVD( n_components = 2 , * , algorithm = 'randomized' , n_iter = 5 , random_state = None , tol = 0.0 )

Снижение размерности с помощью усеченного SVD (также известного как LSA).

Этот преобразователь выполняет уменьшение линейной размерности с помощью усеченного сингулярного разложения (SVD). 
В отличие от PCA, этот оценщик не центрирует данные перед вычислением разложения по сингулярным значениям. 
Это означает, что он может эффективно работать с разреженными матрицами.

```python
svd = TruncatedSVD(n_components = 16)
X_reduced = svd.fit_transform(X)   # fit_transform(X [, y]) - подогнать модель к X и выполнить уменьшение размерности на X
svd.explained_variance_ratio_.sum()
```

```python
rf_reduced = RandomForestClassifier(oob_score = True)
rf_reduced.fit(X_reduced, y) # fit(X [, y]) - подобрать модель на тренировочных данных X
rf_reduced.oob_score_
```

```python
for k in range(5, 100, 20):
    D1 = D.copy()
    D1[D1 < d[int(k)]] = 0
    print(int(k))
    x1 = np.dot(np.dot(U, D1), Vt)
    pl.imshow(x1, cmap = 'gray')
    pl.show()
```
