```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# создаем матрицу 3х3 с числами от 1 до 9
a = np.arange(1,10)
A = np.reshape(a, (3, 3), order = 'C')
A
```

```python
# создаем единичную матрицу
B = np.eye(3, dtype = np.int8)
B
```

```python
# сложим матрицы А и В (А+В = В+А)
C = A + B
C
```

```python
# зададим вектор (1, 2, 3)
v = np.array([1, 2, 3])  # вектор-строка или матрица, размером 1х3
V = np.reshape(v, (3, 1))  # преобразуем в вектор-столбец или матрицу, размером 3х1
V
```

```python
# выполним умножение матрицы C на вектор V
# правило перемножения матриц - число столбцов матрицы С расно числу строк вектора V
R = C @ V
R
```

```python
# посчитаем обратную матрицу к R
# поскольку матрица R не квадратная, обратной матрицы к ней не существует, получим ошибку ...
U = np.linalg.inv(R)
U
```

```python
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y, c = 'b')
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
# Уменьшить объект в два раза - операция деления матрицы А на 2
A_2 = A / 2
A_2
x_2 = A_2[:,0]
y_2 = A_2[:,1]

plt.plot(x_2, y_2)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

### Пространственные преобразования

Матрицы поворота вокруг осей координат имеют вид:


двумерное пространство




трехмерное пространство




четырехмерное пространство



```python
# Повернуть на 130 градусов матрицу А
# зададим матрицу поворота M

M = np.array([[np.cos(13*np.pi/18), np.sin(13*np.pi/18), 0], [- np.sin(13*np.pi/18), np.cos(13*np.pi/18), 0], [0, 0, 1]])
```

```python
S = A @ M
S
```

```python
x_3 = S[:,0]
y_3 = S[:,1]
#z_3 = [:,2]

plt.plot(x_3, y_3)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
# Отразить объект относительно прямой y=x
# с помощью матрицы N [[0, 1, 0], [1, 0, 0], [0, 0, 1]]
N = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
N
```

```python
# линейное отображение (умножение на матрицу)
SS = A @ N
SS
```

```python
x_4 = SS[:,0]
y_4 = SS[:,1]
#z_4 = SS[:,2]


plt.plot(x_4, y_4, c = 'g')
plt.fill_between(x_4, y_4)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```
